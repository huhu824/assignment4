```python
# taxi problem
taxi = gym.make('Taxi-v3', render_mode=None)
```

```python
# policy iteration
v_pi_taxi, v_track_pi_taxi, pi_pi_taxi = Planner(taxi.env.P).policy_iteration(gamma = 0.99, theta=1e-10)
```

```python
# Q-learning
Q_taxi, V_q_taxi, pi_q_taxi, Q_track_taxi, pi_track_taxi = RL(taxi.env).q_learning(gamma=.99, init_alpha=0.5, min_alpha=0.01, alpha_decay_ratio=0.5, 
	init_epsilon=1.0, min_epsilon=0.1, epsilon_decay_ratio=0.9, n_episodes=10000)
```

```python
# compare normalized value matrices of three methods
v_pi_taxi_new = v_pi_taxi/v_pi_taxi.sum()
v_vi_taxi_new = v_vi_taxi/v_vi_taxi.sum()
V_q_taxi_new = V_q_taxi/V_q_taxi.sum()
diff_1 = v_pi_taxi_new-v_vi_taxi_new
diff_2 = V_q_taxi_new-v_vi_taxi_new
print(diff_1, diff_2)
```
